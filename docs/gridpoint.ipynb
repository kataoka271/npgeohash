```python
import matplotlib.pyplot as plt
from matplotlib import patches

try:
    import npgeohash.npgeohash as npgeohash
except ImportError:
    import sys

    sys.path.append("..")

    import npgeohash.npgeohash as npgeohash


def show_gridpoint():
    a, b, r, w, h = 0.5, 0.5, 3000, 1217.1555987614984, 611.49622628141
    ax = plt.subplot()
    ellipse = patches.Ellipse(xy=(a, b), width=2 * r / w, height=2 * r / h, fill=False)
    ax.grid()
    ax.add_patch(ellipse)
    pts = list(npgeohash._gridpoints(a, b, r, w, h))
    print(f"len(pts) = {len(pts)}, len(set(pts)) = {len(set(pts))}")
    x, y = zip(*pts)
    ax.scatter(x, y)


show_gridpoint()
```
